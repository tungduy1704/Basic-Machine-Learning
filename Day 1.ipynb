# Data Types
Data types can be split into three main categories:
- Numerical
- Categorical
- Ordinal

**Numerical** data are numbers, including discrete data & continuous data.

**Categorical** data are values cannot be measured up against each other. Example: color, yes/no.

**Ordinal** data are like categorical data, but can be measured up against each other. Example: school grade A is better than grade B.


# Mean, Median, and Mode
- Mean - the average value.
- Median - the mid point value.
- Mode - the most common value.

```python
import numpy as np
from scipy import stats

speed = [99,86,87,88,111,86,103,87,94,78,77,85,86]

mean = (99+86+87+88+111+86+103+87+94+78+77+85+86) / 13

x_mean = np.mean(speed)

print(x_mean)

x_median = np.median(speed)

print(x_median)

x_mode = stats.mode(speed)
print(x_mode)


```

# Standard Deviation

A low standard deviation means that most of the numbers are close to the mean (average) value.

A high standard deviation means that the values are spread out over a wider range.

```python
import numpy as np

speed = [86,87,88,86,87,85,86]

x_std = np.std(speed)

print(x_std)
```

# Variance
Variance is another number that indicates how spread out the values are.

In fact, if you take the square root of the variance, you get the standard deviation!

Or the other way around, if you multiply the standard deviation by itself, you get the variance!

```python
speed = [32,111,138,28,59,77,97]

x_var = np.var(speed)

print(x_var)
```

# Regression
- Linear regression: uses the relationship between the data-points to draw a straight line through all them

```python
x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]

plt.scatter(x, y)
plt.show()
```

```python
from scipy import stats

slope, intercept, r, p, std_err = stats.linregress(x, y)

def myfunc(x):
  return slope * x + intercept

mymodel = list(map(myfunc, x))

plt.scatter(x, y)
plt.plot(x, mymodel)
plt.show()
```

# R for Relationship

The value r from the above method mean the coefficient of correlation

It ranges from -1 to 1, where 0 means no relationship, and 1 (-1) means 100% related

```python
print(r)
```

# Multiple Regression
It is like the Linear regression, but in this case we will have more than 1 variable

```python
import pandas as pd
from sklearn import linear_model

df = pd.read_csv("data.csv")

X = df[['Weight', 'Volume']]
y = df[['CO2']]

regr = linear_model.LinearRegression() # Use LinearRegression to create a linear regression object
regr.fit(X, y) # This object has an method fit() that takes the independent and dependent values as parameters and fills the regression object with data that describes the relationship

predictedCO2 = regr.predict([[2300, 1300]])
print(predictedCO2)


print(regr.coef_)
```

# Scale value

There are different methods for scaling data, we will use standardization.
$$ z = \frac{x - u}{s} $$
where z is the new value, x is the orginial value, u is the mean and s is the standard deviation.

Python sklearn also offers a method called `StandardScaler()`


```python
import pandas as pd
from sklearn import linear_model
from sklearn.preprocessing import StandardScaler

scale = StandardScaler()

df = pd.read_csv('data.csv')

X = df[['Weight', 'Volume']]
y = df[['CO2']]

scaleX = scale.fit_transform(X)

regr = linear_model.LinearRegression()
regr.fit(scaleX, y)

scaled = scale.transform([[2300, 1.3]])

predictedCO2 = regr.predict([scaled[0]])

print(predictedCO2)

```

# Train - Test
To measure if the model is good enough, we can use a method called train/test.

It is called train/test because you split the data into two sets:
- a training set 
- a testing set


```python
import numpy
import matplotlib.pyplot as plt
numpy.random.seed(2)

x = numpy.random.normal(3, 1, 100)
y = numpy.random.normal(150, 40, 100) / x

plt.scatter(x, y)
plt.show()
```

```python
train_x = x[:80]
train_y = y[:80]

test_x = x[80:]
test_y = y[80:]

plt.scatter(train_x, train_y)
plt.scatter(test_x, test_y)
```

```python
mymodel = numpy.poly1d(numpy.polyfit(train_x, train_y, 5))

myline = numpy.linspace(0, 6, 100)

plt.scatter(train_x, train_y)
plt.plot(myline, mymodel(myline))
plt.show()
```

```python
import numpy
from sklearn.metrics import r2_score
numpy.random.seed(2)

x = numpy.random.normal(3, 1, 100)
y = numpy.random.normal(150, 40, 100) / x

train_x = x[:80]
train_y = y[:80]

test_x = x[80:]
test_y = y[80:]

mymodel = numpy.poly1d(numpy.polyfit(train_x, train_y, 5))

r2 = r2_score(train_y, mymodel(train_x))

print(r2)
```
